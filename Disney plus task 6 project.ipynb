```python
df.type.value_counts().plot(kind='bar',figsize=(6,5), color=('olive','pink','blue','yellow','violet','seagreen','orange','salmon'))
plt.xlabel('type')
plt.ylabel("type of movies")
plt.title("Number of movie types")
plt.show()
```

```python
#2 no of show realse year
df.release_year.value_counts().plot(kind='bar',figsize=(26,10))
plt.xlabel('relase year')
plt.ylabel("Number of show")
plt.title("year")
plt.xticks(rotation=90)
plt.show()
```

```python
#4 release by runtime
df.groupby('release_year')['duration'].value_counts()
```

```python
df.groupby('release_year')['duration'].nunique().plot(kind='bar',figsize=(32,10))
plt.title(" Movies of Varying runtime VS Year")
plt.xlabel("release_year")
plt.ylabel("Count of Movies with different runtime")
plt.show()
```
